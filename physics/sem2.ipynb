```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
A = 3
k = 5
w = 3
X = np.linspace(0, 2*np.pi/k,200)
f = lambda x,t: 1/(1-A*k*np.sin(k*x)*np.cos(w*t))
g = lambda x,t: A*np.cos(k*x)*np.cos(w*t)
```

```python
plt.plot(X, f(X,0), '--b')
plt.plot(X, g(X,0), 'r')
plt.plot(X, f(X,2*np.pi/k), '--y')
plt.plot(X, g(X,2*np.pi/k), 'g')
```
