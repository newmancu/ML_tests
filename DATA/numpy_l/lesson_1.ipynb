```python
from numpy import pi
def f(x,n=100):
    sum = 0
    for i in range(1,n+1):
        sum += (1 - (-1)^i)/(pi*i)*np.sin(i*x)
    return .5 + sum
```

```python

X = np.linspace(-2*pi, 2*pi, 100)
Y = f(X, 10)


plt.plot(X,Y)
```
